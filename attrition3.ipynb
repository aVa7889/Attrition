```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
department_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop=None)

department_encoder.fit(y_train[['Department']])

y_department_train = department_encoder.transform(y_train[['Department']])
y_department_test = department_encoder.transform(y_test[['Department']])

y_department_train, y_department_test

```

```python
attrition_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop=None)

attrition_encoder.fit(y_train[['Attrition']])

y_attr_train = attrition_encoder.transform(y_train[['Attrition']])
y_attr_test = attrition_encoder.transform(y_test[['Attrition']])

y_attr_train, y_attr_test
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)
# Scale the training and testing data

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Create, Compile, and Train the Model

```python
X_df.shape
```

```python
# Create the shared layers of the model

# Input layer
input_layer = layers.Input(shape=(X_df.shape[1],), name='input_features')

# Shared hidden layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Branch for department prediction
y_department_output = layers.Dense(3, activation='softmax', name='y_department_output')(shared_layer2)

# Branch for attrition prediction
y_attr_output = layers.Dense(2, activation='sigmoid', name='y_attr_output')(shared_layer2)
```

```python
y_attr_output
```

```python
keras.utils.set_random_seed(812)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[
    y_department_output, 
    y_attr_output
    ])

# Compile the model
model.compile(optimizer='adam',
            loss={'y_department_output': 'categorical_crossentropy', 'y_attr_output': 'binary_crossentropy'},
            metrics={'y_department_output': 'accuracy', 'y_attr_output': 'accuracy'})

# Display the model summary
model.summary()
```

```python
# Fit the model
model.fit(
    X_train_scaled,
    {
        'y_department_output': y_department_train, 
        'y_attr_output': y_attr_train
    },
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'y_department_output': y_department_test, 'y_attr_output': y_attr_test})
test_results
```

```python
# Print the quality and color accuracy
print(f"Attrition Accuracy: {test_results[1]}")
print(f"Department Accuracy: {test_results[2]}")
```
